```python
# josn 데이터 여러 개 불러오기
import os
import json

path = '/content'

# 경로 내에 파일을 모두 불러옴
file_list = os.listdir(path)

# json 확장자를 가진 파일로 새로운 리스트 생성
json_file_list = [file for file in file_list if file.endswith('.json')]

# josn_file_list에는 json 파일들만 남아있게 된다.
```

```python
import json
import os

df = []  # 데이터를 저장할 리스트 생성
error_files = []  # 오류가 발생한 파일을 저장할 리스트

# json 파일을 하나씩 읽어서 리스트에 추가
for json_file in json_file_list:
    file_path = os.path.join(path, json_file)  # 파일의 전체 경로
    try:
        with open(file_path, "r", encoding="utf-8") as f:  # 인코딩 명시
            data = json.load(f)  # json 데이터 읽어오기  # 각 json 데이터를 리스트에 추가
            for info in data['info']:
               subject = info['annotations']['subject']
               speaker_type = info['annotations']['speaker_type']
               text = info['annotations']['text']
               df.append({'subject': subject, 'speaker_type': speaker_type, 'text': text})
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON in file: {json_file} - {e}")
        error_files.append(json_file)  # 오류 발생 파일 리스트에 추가
    except Exception as e:
        print(f"Unexpected error in file: {json_file} - {e}")
        error_files.append(json_file)

if error_files:
    print("Files with errors:", error_files)
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
import torch
from sklearn.model_selection import train_test_split

texts = df['text'].values
labels = df['subject'].values

# 2. 레이블을 숫자로 변환
label_mapping = {label: idx for idx, label in enumerate(set(labels))}
labels = [label_mapping[label] for label in labels]

# 3. 데이터셋 분할
train_texts, val_texts, train_labels, val_labels = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
# 4. 모델 불러오기
tokenizer = AutoTokenizer.from_pretrained("klue/roberta-small")
num_labels = len(label_mapping)
model = AutoModelForSequenceClassification.from_pretrained("klue/roberta-small", num_labels = num_labels)
```

```python
# 5. 토큰화
train_encodings = tokenizer(list(train_texts), truncation=True, padding=True, max_length=128)
val_encodings = tokenizer(list(val_texts), truncation=True, padding=True, max_length=128)
```

```python
# 6. 데이터셋 준비
class TextDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = TextDataset(train_encodings, train_labels)
val_dataset = TextDataset(val_encodings, val_labels)
```

```python
# 7. 학습 설정
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    evaluation_strategy="epoch",
    report_to="none",
    logging_steps=10,
    save_strategy="epoch",
    load_best_model_at_end=True
)
```

```python
# 8. Trainer 설정
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)
```

```python
trainer.train()
```

```python
# 10. 평가
eval_results = trainer.evaluate()
print("평가 결과:", eval_results)
```

```python
# 11. 새로운 텍스트 예측
test_text = ["중앙은행이 기준금리 인상을 논의하고있다"]
test_encodings = tokenizer(test_text, truncation=True, padding=True, max_length=128, return_tensors="pt")
outputs = model(**test_encodings)

# 예측된 레이블 인덱스 가져오기
predictions = torch.argmax(outputs.logits, dim=-1)

# 숫자 레이블을 원래 카테고리 이름으로 변환하기 위해 역 매핑 생성
inverse_label_mapping = {v: k for k, v in label_mapping.items()}
predicted_label = inverse_label_mapping[predictions.item()]  # 예측된 카테고리 이름

print("예측 결과:", predicted_label)
```
